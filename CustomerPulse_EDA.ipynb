```python
df = df.drop_duplicates(subset=['InvoiceNo','StockCode'])
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head(5)
```

```python
reference_date = df['InvoiceDate'].max()
```

```python
reference_date
```

```python
rfm = df.groupby('CustomerID').agg(
    Recency = ('InvoiceDate','max'),
    Frequency = ('InvoiceNo','count'),
    Monetary = ('TotalPrice','sum')
)
```

```python
rfm.head()
```

```python
rfm['Recency'] = (reference_date - rfm['Recency']).dt.days
```

```python
rfm.head()
```

```python
rfm.shape
```

```python
rfm.describe()
```

```python
df.dtypes
```

```python
rfm.index = rfm.index.astype(int)
```

```python
rfm.head()
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=4, labels=[4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], q=4, labels=[1,2,3,4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], q=4, labels=[1,2,3,4])
```

```python
rfm.head()
```

```python
rfm.dtypes
```

```python
rfm['R_Score'] = rfm['R_Score'].astype(int)
rfm['F_Score'] = rfm['F_Score'].astype(int)
rfm['M_Score'] = rfm['M_Score'].astype(int)
```

```python
rfm['RFM_Score'] = rfm['R_Score']+rfm['F_Score']+rfm['M_Score']
```

```python
rfm.head()
```

```python
rfm['Segment'] = pd.cut(rfm['RFM_Score'], bins=[2,4,6,9,12], labels=['Lost','At Risk','Loyal','Champions'])
```

```python
rfm.head()
```

```python
rfm['Segment'].value_counts()
```

```python
rfm.to_sql('rfm_scores', engine, if_exists='replace', index=True)
```

```python
rfm['Churned'] = (rfm['Recency']>90).astype(int)
```

```python
rfm['Churned'].value_counts()
```

```python
rfm.to_sql('rfm_scores',engine, if_exists='replace', index=True)
```

```python
rfm.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
x = rfm[['R_Score','F_Score','M_Score']]
y = rfm['Churned']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = RandomForestClassifier(class_weight='balanced',random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
print('Training Accuracy',model.score(x_train,y_train))
print('Testing Accuracy',model.score(x_test,y_test))
```

```python
pd.Series(model.feature_importances_, index = ['R_Score','F_Score','M_Score'])
```

```python
import joblib

joblib.dump(model,'customer_pulse_churn_model.pkl')
```
